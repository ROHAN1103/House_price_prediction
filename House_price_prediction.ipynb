```python
#importing necessary librires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading and displayinf dataset
dataset=pd.read_csv('London Property Listings Dataset.csv')
dataset.sample(5)
```

```python
# Display count of missing values per column
print(dataset.isnull().sum())
```

```python
# Check if there are any missing values
print(dataset.isnull().values.any())
```

```python
# Heatmap to visualize missing values
sns.heatmap(dataset.isnull(), cbar=False, cmap='viridis')
plt.show()
```

```python
#copying a dataset to keep original dataset as it is.
df=dataset
df.sample(5)
```

```python
df.describe()
```

```python
df.info()
```

```python
df=df.drop('Postcode', axis='columns')
df.sample(5)
```

```python
df['PropertyType'].value_counts().plot(kind='barh')
df['PropertyType'].value_counts()
```

```python
df['Area'].value_counts().plot(kind='barh')
df['Area'].value_counts()
```

```python
x=df[df['Area']=='Kingston']
print(x)
```

```python
df=df.drop('Area_Avg_Price', axis='columns')
df.sample(5)
```

```python
df=df.drop('Price_Category', axis='columns')
```

```python
df.sample(5)
```

```python
from sklearn.preprocessing import OneHotEncoder
df=pd.get_dummies(df, columns=['PropertyType','Area'])
```

```python
df.sample(5)
```

```python
X=df.drop('Price', axis='columns')
y=df['Price']
```

```python
X.sample(5)
```

```python
for columns in X:
    if X[columns].dtype!=int:
        X[columns]=X[columns].astype(int)
```

```python
X.sample(4)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X[['Bedrooms', 'Bathrooms','Size']]=scaler.fit_transform(X[['Bedrooms', 'Bathrooms','Size']])
```

```python
X.sample(3)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
mse=mean_squared_error(y_test, y_pred)
r2=r2_score(y_test,y_pred)
```

```python
y_pred
```

```python
mse
```

```python
r2
```

```python
from sklearn.tree import DecisionTreeRegressor
model2=DecisionTreeRegressor()
model2.fit(X_train, y_train)
```

```python
y_pred2=model2.predict(X_test)
```

```python
mse2=mean_squared_error(y_test, y_pred2)
r22=r2_score(y_test,y_pred2)
print(mse2)
print(r22)
```
